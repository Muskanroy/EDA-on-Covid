### Post Vaccination breakthrough infection

```python
data1={'name1':['Covaxin','Received 1st Dose','Received 2nd Dose'],
     'Doses1':[11000000,9356436,1737178],
     'positive1':[4963,4208,695]}
data2={'name2':['Covishield','Received 1st Dose','Received 2nd Dose'],
      'Doses2':[116000000,100302745,15732754],
      'positive2':[22159,17145,5014]}
df1=pd.DataFrame.from_dict(data1)
df2=pd.DataFrame.from_dict(data2)
df1
```

The plot below shows the post vaccination infection after 1st and 2nd dose of Covaxin and Covishield

```python
plt.figure(figsize=(16,6))
ax1=plt.subplot(1,1,1)
ax1.bar(df1.name1,df1.Doses1,0.4,color='y',label='Vaccinated')
ax1.yaxis.grid(linestyle='-')
ax1.set_ylabel('vacciantion',fontdict={'fontweight':'bold','fontsize':14})
ax1.legend()
ax3=ax1.twinx()
ax3.plot(df1.name1,df1.positive1,'o-',color='b',label='+ve after vaccination',linewidth=3)
ax1.set_title("Post Vaccination Breakthrough Infection",fontdict={'fontweight':'bold','fontsize':20})
ax3.set_ylabel('+ve after vaccination',fontdict={'fontweight':'bold','fontsize':14})
ax3.legend()
plt.show()
plt.figure(figsize=(16,8))
ax2=plt.subplot(2,1,2)
ax2.bar(df2.name2,df2.Doses2,0.4,color='g',label='Vaccinated')
ax2.set_ylabel('vacciantion',fontdict={'fontweight':'bold','fontsize':14})
ax2.yaxis.grid(linestyle='-')
ax2.legend()
ax4=ax2.twinx()
ax4.plot(df2.name2,df2.positive2,'o-',color='r',label='+ve after vaccination',linewidth =3)

ax4.set_ylabel('+ve after vaccination',fontdict={'fontweight':'bold','fontsize':14})
ax4.legend()
plt.show()
```

## Correlation between vaccine administered, Active cases and Recovered

```python
import pandas as pd
```

```python
#loading file containing daywise vaccine administered
correlation=pd.read_csv('C:\\Users\\muska\\Downloads\\covid_vaccine_statewise.csv')
#changing datatype of column
correlation['Updated On']=pd.to_datetime(correlation['Updated On'],format= '%d/%m/%Y')
#getting data of india
corr=correlation.loc[(correlation['State']=='India') & (correlation['Updated On']<='16/4/2021')]
corr
```

```python
covaxin=corr['Total Covaxin Administered']
covaxin
```

```python
covishield=corr['Total CoviShield Administered']
covishield
```

```python
total=corr['Total Doses Administered']
total
```

```python
#loading file containing daywise cases in India
case =pd.read_csv('C:\\Users\\muska\\Downloads\\COVIDnew.csv')
# changing datatype of column
case['Date'] = pd.to_datetime(case['Date'], format='%d-%m-%Y')
# getting data of India only
covid=case.loc[(case['Region']=='India') & (case['Date']>='16-01-2021')]
covid
```

```python
active=covid['Active Cases']
active
```

```python
recovered=covid['Cured/Discharged']
recovered
```

The below plot shows the correlations between Covaxin,Covishield and Total vaccine administered with active cases and recovered cases 

```python
plt.figure(figsize=(16,8))
plt.scatter(covaxin,active,label='Active cases',alpha=1,s=100)
plt.scatter(covaxin,recovered,label='Recovered',alpha=1,s=100)
plt.xlabel('covaxin administered',fontdict={'fontweight':'bold','fontsize':18})
plt.ylabel('no of cases',fontdict={'fontweight':'bold','fontsize':18})
plt.title("Correlation between covaxin administered with Active cases and recovered cases",fontdict={'fontweight':'bold','fontsize':22})
plt.legend()
plt.grid()
plt.show()
plt.figure(figsize=(16,8))
plt.scatter(covishield,active,label='Active cases',alpha=1,s=100)
plt.scatter(covishield,recovered,label='Recovered',alpha=1,s=100)
plt.xlabel('covishield administered',fontdict={'fontweight':'bold','fontsize':18})
plt.ylabel('no of cases',fontdict={'fontweight':'bold','fontsize':18})
plt.title("Correlation between covishield administered with Active cases and recovered cases",fontdict={'fontweight':'bold','fontsize':22})
plt.legend()
plt.grid()
plt.show()
plt.figure(figsize=(16,8))
plt.scatter(total,active,label='Active cases',alpha=1,s=100)
plt.scatter(total,recovered,label='Recovered',alpha=1,s=100)
plt.xlabel('total vaccine administered',fontdict={'fontweight':'bold','fontsize':18})
plt.ylabel('no of cases',fontdict={'fontweight':'bold','fontsize':18})
plt.title("Correlation between total vaccine administered with Active cases and recovered cases",fontdict={'fontweight':'bold','fontsize':22})
plt.legend()
plt.grid()
plt.show()
```
